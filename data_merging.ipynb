```python

import pandas as pd
```

```python
dumps_13 = pd.read_csv("./data/water_resources/13/3_Водні_ресурси_2013 - water_contam.csv").drop(0)
dumps_13 = dumps_13.rename(columns={dumps_13.columns[0]: "Region"})

dumps_13_no_clean = pd.read_csv("./data/water_resources/13/3_Водні_ресурси_2013 - water_contam_no_clean.csv").drop(0)
dumps_13_no_clean = dumps_13_no_clean.rename(columns={dumps_13_no_clean.columns[0]: "Region"})

dumps_13_not_enough_clean = pd.read_csv(
    "./data/water_resources/13/3_Водні_ресурси_2013 - water_cont_not_enough_clean.csv").drop(0)
dumps_13_not_enough_clean = dumps_13_not_enough_clean.rename(columns={dumps_13_not_enough_clean.columns[0]: "Region"})

clearing_plants_13 = pd.read_csv("./data/water_resources/13/3_Водні_ресурси_2013 - clearing_plants.csv").drop(0)
clearing_plants_13 = clearing_plants_13.rename(columns={clearing_plants_13.columns[0]: "Region"})

rivers_13 = pd.read_csv("./data/water_resources/13/3_Водні_ресурси_2013 - rivers.csv")
rivers_13 = rivers_13.rename(columns={rivers_13.columns[0]: "Region"})
rivers_13.columns = [col if col == "Region" else col.lower() for col in rivers_13.columns]

dumps_13
```

```python
dumps_14_16 = pd.read_csv("./data/water_resources/14-16/4_Водні_ресурси_2014-2016 - water_contam.csv").drop(0)
dumps_14_16 = dumps_14_16.rename(columns={dumps_14_16.columns[0]: "Region"}).drop(dumps_14_16.tail(2).index)

dumps_14_16_no_clean = pd.read_csv(
    "./data/water_resources/14-16/4_Водні_ресурси_2014-2016 - water_contam_not_clean.csv").drop(0)
dumps_14_16_no_clean = dumps_14_16_no_clean.rename(columns={dumps_14_16_no_clean.columns[0]: "Region"}).drop(
    dumps_14_16_no_clean.tail(2).index)

dumps_14_16_not_enough_clean = pd.read_csv(
    "./data/water_resources/14-16/4_Водні_ресурси_2014-2016 - water_contam_not_enough.csv").drop(0)
dumps_14_16_not_enough_clean = dumps_14_16_not_enough_clean.rename(
    columns={dumps_14_16_not_enough_clean.columns[0]: "Region"}).drop(dumps_14_16_not_enough_clean.tail(2).index)

clearing_plants_14_16 = pd.read_csv(
    "./data/water_resources/14-16/4_Водні_ресурси_2014-2016 - cleaning_plants.csv").drop(0)
clearing_plants_14_16 = clearing_plants_14_16.rename(columns={clearing_plants_14_16.columns[0]: "Region"}).drop(
    clearing_plants_14_16.tail(2).index)

rivers_14 = pd.read_csv("./data/water_resources/14-16/3_Водні_ресурси_2014 - rivers.csv")
rivers_14 = rivers_14.rename(columns={rivers_14.columns[0]: "Region"})

rivers_15 = pd.read_csv("./data/water_resources/14-16/3_Водні_ресурси_2015 - rivers.csv")
rivers_15 = rivers_15.rename(columns={rivers_15.columns[0]: "Region"})

rivers_16 = pd.read_csv("./data/water_resources/14-16/4_Водні_ресурси_2016 - rivers.csv")
rivers_16 = rivers_16.rename(columns={rivers_16.columns[0]: "Region"})
```

```python
dumps_17_19 = pd.read_csv("./data/water_resources/17-19/україна_довкілля_2017-2019 - water_contam.csv").drop(0)
dumps_17_19 = dumps_17_19.rename(columns={dumps_17_19.columns[0]: "Region"})

dumps_17_19_no_clean = pd.read_csv(
    "./data/water_resources/17-19/україна_довкілля_2017-2019 - water_contam_no_clean.csv").drop(0)
dumps_17_19_no_clean = dumps_17_19.rename(columns={dumps_17_19_no_clean.columns[0]: "Region"})

dumps_17_19_not_enough_clean = pd.read_csv(
    "./data/water_resources/17-19/україна_довкілля_2017-2019 - water_contam_not_enough_clean.csv").drop(0)
dumps_17_19_not_enough_clean = dumps_17_19_not_enough_clean.rename(
    columns={dumps_17_19_not_enough_clean.columns[0]: "Region"})

clearing_plants_17_19 = pd.read_csv(
    "./data/water_resources/17-19/україна_довкілля_2017-2019 - cleaning_plants.csv").drop(0)
clearing_plants_17_19 = clearing_plants_17_19.rename(columns={clearing_plants_17_19.columns[0]: "Region"})

rivers_17 = pd.read_csv("./data/water_resources/17-19/4_Водні_ресурси_2017 - rivers.csv")
rivers_17 = rivers_17.rename(columns={rivers_17.columns[0]: "Region"})

rivers_18 = pd.read_csv("./data/water_resources/17-19/4_Водні_ресурси_2018 - rivers.csv")
rivers_18 = rivers_18.rename(columns={rivers_18.columns[0]: "Region"})

rivers_19 = pd.read_csv("./data/water_resources/17-19/19 - rivers.csv")
rivers_19 = rivers_19.rename(columns={rivers_19.columns[0]: "Region"})
```

```python
rivers_dataframes = [rivers_13, rivers_14, rivers_15, rivers_16, rivers_17, rivers_18, rivers_19]

for df in rivers_dataframes:
    df["Region"] = df["Region"].str.replace(" ", "")

prefix = "р."


def remove_prefix(region):
    if region.startswith(prefix):
        return region[len(prefix):].strip()
    return region.strip()


regions_with_prefix = {remove_prefix(region) for region in rivers_18["Region"] if region.startswith(prefix)}

rivers_19["Region"] = rivers_19["Region"].apply(
    lambda region: prefix + region if (
            remove_prefix(region) in regions_with_prefix and not region.startswith(prefix)) else region
)

for df in rivers_dataframes:
    df.columns = [col.replace(" ", "") for col in df.columns]

rivers_13.columns
```

```python
dumps_20_22 = pd.read_csv("./data/water_resources/20-22/україна_довкілля_2020-2022 - water_contam.csv").drop(0)
dumps_20_22 = dumps_20_22.rename(columns={dumps_20_22.columns[0]: "Region"}).drop(dumps_20_22.tail(2).index)

dumps_20_22_no_clean = pd.read_csv(
    "./data/water_resources/20-22/україна_довкілля_2020-2022 - water_contam_no_clean.csv").drop(0)
dumps_20_22_no_clean = dumps_20_22_no_clean.rename(columns={dumps_20_22_no_clean.columns[0]: "Region"}).drop(
    dumps_20_22_no_clean.tail(2).index)

dumps_20_22_not_enough_clean = pd.read_csv(
    "./data/water_resources/20-22/україна_довкілля_2020-2022 - water_contam_not_enough.csv").drop(0)
dumps_20_22_not_enough_clean = dumps_20_22_not_enough_clean.rename(
    columns={dumps_20_22_not_enough_clean.columns[0]: "Region"}).drop(dumps_20_22_not_enough_clean.tail(2).index)
```

```python
dumps_dfs = [dumps_13, dumps_14_16, dumps_17_19, dumps_20_22]

melted_dfs = [
    df.melt(id_vars=["Region"], var_name="year", value_name="polluted_dumps")
    for df in dumps_dfs
]

merged_dumps = pd.concat(melted_dfs, ignore_index=True)

merged_dumps["year"] = merged_dumps["year"].str.strip().astype(int)
```

```python
dumps_no_clean_dfs = [dumps_13_no_clean, dumps_14_16_no_clean, dumps_17_19_no_clean, dumps_20_22_no_clean]

melted_no_clean = [
    df.melt(id_vars=["Region"], var_name="year", value_name="not_cleaned_dumps")
    for df in dumps_no_clean_dfs
]

merged_no_clean = pd.concat(melted_no_clean, ignore_index=True)

merged_no_clean["year"] = merged_no_clean["year"].str.strip().astype(int)

merged_dumps = pd.merge(merged_dumps, merged_no_clean, on=["Region", "year"], how="left")
```

```python
dumps_not_enough_clean_dfs = [
    dumps_13_not_enough_clean,
    dumps_14_16_not_enough_clean,
    dumps_17_19_not_enough_clean,
    dumps_20_22_not_enough_clean
]

melted_not_enough = [
    df.melt(id_vars=["Region"], var_name="year", value_name="dumps_not_cleaned_enough")
    for df in dumps_not_enough_clean_dfs
]

merged_not_enough = pd.concat(melted_not_enough, ignore_index=True)

merged_not_enough["year"] = merged_not_enough["year"].str.strip().astype(int)

merged_dumps = pd.merge(merged_dumps, merged_not_enough, on=["Region", "year"], how="left")

clearing_plants_dfs = [clearing_plants_13, clearing_plants_14_16, clearing_plants_17_19]

melted_clearing = [
    df.melt(id_vars=["Region"], var_name="year", value_name="num_clearing_plants")
    for df in clearing_plants_dfs
]

merged_clearing = pd.concat(melted_clearing, ignore_index=True)

merged_clearing["year"] = merged_clearing["year"].str.strip().astype(int)

merged_dumps = pd.merge(merged_dumps, merged_clearing, on=["Region", "year"], how="left")
```

```python
import pandas as pd

river_dfs_with_year = [
    (rivers_13, 2013),
    (rivers_14, 2014),
    (rivers_15, 2015),
    (rivers_16, 2016),
    (rivers_17, 2017),
    (rivers_18, 2018),
    (rivers_19, 2019)
]

for df, year in river_dfs_with_year:
    df["year"] = year

merged_rivers = pd.concat([df for df, _ in river_dfs_with_year], ignore_index=True)
```

```python
merged_dumps.to_csv("./data/water_resources/merged/merged_dumps.csv", index=False)
merged_rivers.to_csv("./data/water_resources/merged/merged_rivers.csv", index=True)
```
